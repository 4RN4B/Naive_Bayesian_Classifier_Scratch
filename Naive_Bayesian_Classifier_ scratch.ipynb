```python
x_test = {}
y_test = {}
x_train = {}
y_train = {}

for i in random80:
    x_train[i] = x_list[random_array[i]]
    y_train[i] = y_list[random_array[i]]

for i in random20:
    x_test[i] = x_list[random_array[i]]
    y_test[i] = y_list[random_array[i]]
```

```python
unique_y_train = set(y_train.values())
y_train_list = list(y_train.values())
class_probability = {}
count_class = {}
for i in unique_y_train:
    count_class[i] = y_train_list.count(i)
    class_probability[i] = round(count_class[i]/len(y_train), 3)
```

```python
def class_determine(feature,column,type_class):
    count=0
    for i in x_train:
      if feature == x_train[i][column] and y_train[i] == type_class:
        count += 1
    return count/count_class[type_class]
```

```python
feature_probability = {}
product = 1
product2 = 0
for i in x_test:
  product2 = 0
  for k in unique_y_train:
      c = 0
      product = 1
      print(i, k)
      for j in x_test[i]:
        # print(j)
        product *= class_determine(j, c, k)
        c += 1 
        #print(product)
      product *= class_probability[k]
      print(product)
      if product2 < product:
        product2 = product
        feature_probability[i] = k
  print('---')
```

```python
unique_value = {}
count = 0
for i in unique_y_train:
    unique_value[i] = count
    count += 1
```

```python
y_predict = []
for i in feature_probability:
    y_predict += [unique_value[feature_probability[i]]]
    #print(unique_value[feature_probability[i]])
y_predict = np.array(y_predict)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print('Confusion_matrix=>')
print(confusion_matrix(y_test_array, y_predict))
print('Classification Report=>')
print(classification_report(y_test_array, y_predict))
```
